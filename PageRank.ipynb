```
from google.colab import drive
from scipy.sparse import csc_matrix
from __future__ import division
from fractions import Fraction
import numpy as np
drive.mount('/content/drive',force_remount=True)
data = open('/content/drive/My Drive/Colab Notebooks/Pagerank/dataset.dat','r').read()
data = [item.split() for item in data.split('\n')[:-1]]
data = np.array(data)
nodes = []
edges = []
limit = data.shape
limit =limit[0]
i=0
while (i<limit):
	if(data[i,0]=='n'):
		nodes.append(data[i,1])
	if(data[i,0]=='e'):
		edges.append([data[i,1],data[i,2]])
	i=i+1
```

```
matrix=[]
row=[]
nodecount=len(nodes)
edgecount = len(edges)
edges=np.array(edges)
for i in range(0,nodecount):
	for j in range (0,nodecount):
		row.append(0)
	matrix.append(row)
	row=[]
matrix=np.array(matrix)
```

```
c=0
while (c<edgecount):
  i=int(edges[c,0])
  j=int(edges[c,1])
  matrix[i,j]=1
  c=c+1
```

```
sums=matrix.sum(axis=1)
c=0
while (c<nodecount):
  if (sums[c]==0):
    sums[c]=1
  matrix[c]=[Fraction(i/sums[c]) for i in matrix[c]]
  c=c+1
```

```
def pageRank(G,d=0.85):
  prob = Fraction(1/nodecount)
  dp = np.zeros((nodecount,nodecount))
  dp[:]=prob
  M = d*G+(1-d)*dp
  r = np.matrix([prob for i in range(0,nodecount)])
  r = np.transpose(r)
  r0 = r
  for i in range(0,100):
    r = M * r
    print np.transpose(r)
    if(r0==r).all():
      break
    r0=r
  return r
```
